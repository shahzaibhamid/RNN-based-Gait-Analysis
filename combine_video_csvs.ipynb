```python
import pandas as pd
import pickle
import numpy as np
```

```python
alldata_processed = pd.read_csv("C:/PHD Work/processed/alldata_processed.csv")
video_ids = set(alldata_processed['videoid'])
```

```python
#read all the processed videos into memory (a big dictionary)
all_videos = {}
for i,video_id in enumerate(video_ids):
    print(i,video_id)
    if(i % 100 == 0):
        print("%.2f percent done" % (i*100./len(video_ids)))
    file_string = 'C:/PHD Work/processed/video_csvs/%s.csv' % (video_id)
    print(file_string)
    try: 
        all_videos[video_id] = np.loadtxt(file_string, delimiter=',')
    except:
        print("missing %s" % (video_id))
        #NOTHING IS ENTERING IN all_videos
```

```python
with open('C:/PHD Work/processed/all_videos_dict.pickle', 'wb') as handle:
    pickle.dump(all_videos, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
